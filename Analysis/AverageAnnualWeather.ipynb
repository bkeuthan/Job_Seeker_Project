```python
import matplotlib.pyplot as plt
from itertools import cycle, islice
import pandas as pd
import numpy as np
```

```python
Weather_Average = "Downloads\Weather Average.csv"
```

```python
Weather_ranks = pd.read_csv(Weather_Average)
Weather_ranks.head()
```

```python
Annual_Average_df = Weather_ranks
Annual_Average_df
```

```python
Annual_Average_df['mean'] = Annual_Average_df.iloc[:, 0:49].mean(axis=1)
Annual_Average_df

Annual_Average_df2 = Annual_Average_df
Annual_Average_df2
```

```python
Annual_Average_df["Rank Coldest to Warmest"] = Annual_Average_df2["mean"].rank(method='min', ascending=True)
Annual_Average_df2.sort_values("mean", inplace = True)
Annual_Average_df2
```

```python
Average_Weather_Ranking_df = Annual_Average_df2.drop(columns=['2018', '2019', '2020', '2021', '2022'])
Average_Weather_Ranking_df
```

```python
Annual_Average_df2.to_csv("AverageAnnualWeather.csv", encoding='utf-8', index=False)
```

```python
AverageAnnualWeather = Annual_Average_df2.plot(x='State', y='Rank Coldest to Warmest', kind = "bar", width = 0.7, rot=90, fontsize=6)

plt.title('Annual Average Weather by State', fontweight = 'bold', fontsize=12)
plt.xlabel('States', fontweight = 'bold', fontsize=8)  
plt.ylabel('Temperature', fontweight = 'bold', fontsize=8)

plt.show()
```
