```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
States_Data_Rankings = "Downloads\State_Data_Rankings.csv"
Overall_Ranks = "Downloads\States.csv"
```

```python
State_ranks = pd.read_csv(Overall_Ranks)
```

```python
States_Data = pd.read_csv(States_Data_Rankings)
States_Data.head()
```

```python
crime_rank_df = States_Data.drop(columns=['Affordability', 'Economy', 'Education', 'Healthcare', 'Infrastructure', 'Quality of Life', 'Opportunities'])
crime_rank_df.head()
```

```python
crime_rank_df2 = pd.merge(State_ranks, crime_rank_df, how="left", on=["Overall Rank"])
crime_rank_df2
```

```python
crime_rank_df2["Crime Ranks"] = crime_rank_df2["Crime"].rank(method='min', ascending=True)
crime_rank_df2.sort_values("Crime", inplace = True)
crime_rank_df2
```

```python
crime_rank_df2.to_csv("CrimeRank.csv", index=False, header=True)
```

```python
CrimeRank = crime_rank_df2.plot(x='State', y='Crime', kind = "bar", width = 0.7, rot=90, fontsize=6)

plt.title('Crime Safety Ranks by State', fontweight = 'bold', fontsize=12)
plt.xlabel('States', fontweight = 'bold', fontsize=8)  
plt.ylabel('Rankings', fontweight = 'bold', fontsize=8)

plt.show()
```

```python
CrimeSafetyRanksbyState = CrimeRank
CrimeSafetyRanksbyState.get_figure().savefig('CrimeRank.pdf', format='pdf', bbox_inches='tight')
```
