```python
import pandas as pd
```

```python
edu_df = pd.read_csv('Edu_Data_By_State.csv')
```

```python
print(edu_df)
```

```python
print(edu_df.dtypes)
```

```python
edu_df['Total Score'] = edu_df['Student Success'] + edu_df['Student Safety'] + edu_df['School Quality']

print(edu_df)
```

```python
edu_df = edu_df.drop('RANK', axis=1)

print(edu_df)
```

```python
edu_df['Total Score Rank'] = edu_df['Total Score'].rank(method='max', ascending=True)

print(edu_df)
```

```python
edu_df = edu_df.sort_values(by='Total Score', ascending=True)

print(edu_df)
```

```python
# export the DataFrame to a CSV file
edu_df.to_csv('edu_ranks.csv', index=False)
```

```python
print(edu_df.dtypes)
```

```python
import matplotlib.pyplot as plt

# sort by Total Score Rank
edu_df = edu_df.sort_values('Total Score Rank')

# create plot
fig, ax = plt.subplots(figsize=(16, 8))

# plot bars
ax.bar(edu_df['State'], edu_df['Student Success'], color='b', alpha=0.5, label='Student Success')
ax.bar(edu_df['State'], edu_df['Student Safety'], color='g', alpha=0.5, label='Student Safety')
ax.bar(edu_df['State'], edu_df['School Quality'], color='r', alpha=0.5, label='School Quality')

# Add xticks on the middle of the group bars
ax.set_xlabel('State', fontweight='bold')
ax.set_ylabel('Score', fontweight='bold')
ax.set_xticklabels(edu_df['State'], rotation=90)

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# calculate total score
edu_df['Total Score'] = edu_df['Student Success'] + edu_df['Student Safety'] + edu_df['School Quality']

# create plot
fig, ax = plt.subplots(figsize=(16, 8))

# plot bars for each column
ax.bar(edu_df['State'], edu_df['Student Success'], color='b', alpha=0.5, label='Student Success')
ax.bar(edu_df['State'], edu_df['Student Safety'], color='g', alpha=0.5, label='Student Safety', bottom=edu_df['Student Success'])
ax.bar(edu_df['State'], edu_df['School Quality'], color='r', alpha=0.5, label='School Quality', bottom=edu_df['Student Success']+edu_df['Student Safety'])

# Add xticks on the middle of the group bars
ax.set_xlabel('State', fontweight='bold')
ax.set_ylabel('Score', fontweight='bold')
ax.set_xticklabels(edu_df['State'], rotation=90)

# Add legend
ax.legend()

# Show plot
plt.show()

# save the plot as a PNG file with higher resolution
plt.savefig('myplot.png', dpi=300)
```
