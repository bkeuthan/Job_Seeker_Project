```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
#read csv
CostOfLiving_Rank = pd.read_csv("../output/CostOfLiving_Rank.csv")

CostOfLiving_Rank
```

# Graphs

# State & Local Tax Ranking

```python
StateLocalTaxRanking = CostOfLiving_Rank.plot(x='state', y='StateLocalTaxRank', kind = "bar", width = 0.7, rot=90, fontsize=5)

plt.title('State & Local Tax Ranking', fontweight = 'bold', fontsize=15)
plt.xlabel('States', fontweight = 'bold', fontsize=8)  
plt.ylabel('Ranking', fontweight = 'bold', fontsize=8)

plt.show()
plt.close()
```

```python
StateLocalTaxRanking.get_figure().savefig('StateLocalTaxRanking.pdf', format='pdf', bbox_inches='tight')
```

# Income Tax Ranking 

```python
IncomeTaxRanking = CostOfLiving_Rank.sort_values(by='IncomeTaxRank', ascending=True)

IncomeTaxRanking.plot(x='state', y='IncomeTaxRank', kind = "bar", width = 0.7, rot=90, fontsize=5)

plt.title('Income Tax Ranking', fontweight = 'bold', fontsize=15)
plt.xlabel('States', fontweight = 'bold', fontsize=8)  
plt.ylabel('Ranking', fontweight = 'bold', fontsize=8)

plt.show()
```

```python
IncomeTaxRanking.get_figure().savefig('IncomeTaxRanking.pdf', format='pdf', bbox_inches='tight')
```
