```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load
healthcare_csv = "../data/healthcare.csv"

#read csv
healthcare = pd.read_csv(healthcare_csv)

healthcare.head()
```

```python
#Calculate Total Score 

healthcare['TotalScore'] = healthcare['hospital quality'] + healthcare['preventable admissions'] + healthcare['preventable admissions'] + healthcare['medicare quality']

#Average Rank Score
healthcare["healthcareRank"] = healthcare["TotalScore"].rank(method='min', ascending=True)

healthcare
```

```python
#export data
healthcare.to_csv("../output/healthcareRank.csv", index=False, header=True)
```
